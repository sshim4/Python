**5.Histogram**

```python
Hist_x = [1, 1, 1, 5, 5, 5, 5, 8, 8, 8, 8, 9, 9, 9, 10, 10, 15, 15, 20, 21, 22, 23, 30, 40, 40, 40, 45, 45, 45, 50, 50]
```

```python
plt.rcParams["figure.figsize"] = (10,5)
plt.hist(Hist_x, bins=3,  ec='pink')
plt.show()
```

```python
# histogram with bins=7
plt.hist(Hist_x, bins=7, ec='pink')
plt.show()
```

```python
# Draw two graph in a single plots
figure, axis = plt.subplots(1, 2)

axis[0].hist(Hist_x, bins=3, ec='pink', alpha= 0.5)
axis[0].set_title('Histogram with bins=3')
axis[1].hist(Hist_x, bins=7, ec='pink')
axis[1].set_title('Histogram with bins=7')

plt.show()
```

**6.Customization**

```python
def fun_1(x_value):
    result = x**2 + 4*x + 4
    return result
```

```python
x = np.linspace(0, 10, 11)
y = fun_1(x)
```

```python
print(x)
print(y)
```

```python
# basic customization
plt.scatter(x, y, alpha=0.7)
xlab = 'x value'
ylab = 'y value'
title = '$x^2 + 4x + 4 $'

# Add axis labels
plt.xlabel(xlab)
plt.ylabel(ylab)

# Add title
plt.title(title)

# Display the plot
plt.show()
```

```python
# compare two customized plots

# Draw two graph in a single plots
figure, axis = plt.subplots(1, 2)
figure.tight_layout(pad=5.0) # spacing between two subplots


# Plot1 #################################
axis[0].scatter(x, y, alpha=0.7)
axis[0].set_title('$x^2 + 4x + 4 $')
axis[0].set_xlabel('x value')
axis[0].set_ylabel('y value')

# Plot2 #################################

size = y*2 
axis[1].scatter(x, y, 
                alpha=0.7, 
                s = size, 
                c=['red', 'red', 'blue', 'blue', 'black', 'black', 'black', 'black', 'pink', 'pink', 'pink']
               )
axis[1].set_title('$x^2 + 4x + 4 $')
axis[1].set_xlabel('x value')
axis[1].set_ylabel('y value')


# x_axis and x_axis_lab
x_axis = np.linspace(0, 10, 11)
x_axis_lab = ['0MM', '1MM', '2MM', '3MM', '4MM', '5MM', '6MM', '7MM', '8MM', '9MM', '10MM']

axis[1].set_xticks(x_axis, x_axis_lab, rotation = 45)
# or 
# plt.xticks(x_axis,x_axis_lab, rotation = 45)


# #y_axis and y_axis_lab
y_axis = [0, 40, 80, 120, 160]
y_axis_lab = ['0Kg', '40Kg', '80Kg', '120Kg','160Kg']
axis[1].set_yticks(y_axis, y_axis_lab, rotation = 45)
# plt.yticks(y_axis, y_axis_lab, rotation = 45)

# index data
plt.text(3, 40, 'Here', color='red')

# Add grid() call
plt.grid(True)

########################################

plt.show()
```

**8. 3D scatterplot**

```python
def fun_2(x_value):
    result1 = x**2 + 4*x -3
    result2 = result1 * 2
    return result1, result2
```

```python
x = np.linspace(0, 2, 25)
output = fun_2(x)
```

```python
y = output[0]
z = output[1]
```

```python
# Plot
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.scatter(x, y, z)

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
```
